```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,20)
plt.rcParams["font.family"] = "sans-serif"
CURRENT_HANAU = "Hanau03"
DATASET_DIR = "data_understanding/dataset_results/{}.csv".format(CURRENT_HANAU)
DATASET_DIR = r"D:\Projects\shared_repo\data_understanding\metadata\dataset_results\Hanau02_03.csv"
print(DATASET_DIR)
```

```python
df = pd.read_csv(DATASET_DIR, index_col=0)

df = df[["Duration", "Label", "Type"]]
df
```

```python
temp_df = df.loc[df["Type"] == "speaker_aggressiveness"]
temp_df = temp_df.groupby(["Label"]).sum().sort_values(by="Duration",ascending=False)
temp_df["Duration"] = (temp_df["Duration"] / 60) /60
temp_df
ax = temp_df.plot(kind='bar', title='Speaker Aggressiveness Labels',
               ylabel='Duration (Hours)', figsize=(20, 10), legend = False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')

ax.tick_params(bottom=False, left=False)

ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)
ax.set_title('Speaker Aggressiveness Labels', fontsize=25)
ax.set_ylabel('Duration (Hours)', fontsize=18)
plt.xticks(rotation=0, fontsize = 18)
plt.xlabel("")
ax.figure.savefig("images/speaker_agg_labels.jpeg".format(CURRENT_HANAU), bbox_inches="tight")
#plt.savefig("images/{}/speaker_agg_labels.png".format(CURRENT_HANAU))
```

```python
temp_df = df.loc[df["Type"] == "speaker_event"]
temp_df = temp_df.groupby(["Label"]).sum().sort_values(by="Duration", ascending=False)
temp_df["Duration"] = (temp_df["Duration"] / 60)
temp_df
ax = temp_df.plot(kind='bar', title='Speaker Events',
               ylabel='Duration (Minutes)', figsize=(20, 10), legend = False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')

# Second, remove the ticks as well.
ax.tick_params(bottom=False, left=False)

# Third, add a horizontal grid (but keep the vertical grid hidden).
# Color the lines a light gray as well.
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)
ax.set_title('Speaker Events', fontsize=25)
ax.set_ylabel('Duration (Minutes)', fontsize=18)
plt.xticks(rotation=0, fontsize = 14)
plt.xlabel("")
plt.savefig("images/speaker_events.jpeg", bbox_inches="tight")
```
