```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
RANDOM_FILE = r"D:\Projects\shared_repo\data_understanding\audio_files\audio_input_plots\blues.wav"
```

```python
y, sr = librosa.load(RANDOM_FILE)
```

```python
def extract_cqt(y, sr):
    cqt = np.abs(librosa.cqt(y, sr=sr))
    cqt = librosa.amplitude_to_db(cqt)
    return cqt
```

```python
def extract_chromagram(y, sr):
    chroma = np.abs(librosa.feature.chroma_stft(y=y, sr=sr))
    chroma = librosa.amplitude_to_db(chroma)
    return chroma
    
```

```python
cqt = extract_cqt(y, sr)
librosa.display.specshow(cqt, sr=sr, x_axis="time", y_axis="cqt_note")
plt.colorbar(format="%+2.f dB")
plt.savefig("images/cqt.png", dpi = 100)
```

```python
chroma = extract_chromagram(y, sr)
librosa.display.specshow(chroma, sr=sr, x_axis="time", y_axis="chroma")
plt.colorbar(format="%+2.f dB")
plt.savefig("images/chroma.png", dpi = 100)
```
