```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
plt.rcParams["figure.figsize"] = (20,10)
plt.rcParams["font.family"] = "sans-serif"
DATASET_DIR = r"D:\Projects\shared_repo\data_preparation\metadata_preprocessing\results\dataset_san_3win_3step.csv"
print(DATASET_DIR)
```

```python
df = pd.read_csv(DATASET_DIR)
```

```python
# duration_df = df.loc[df["Duration"] > 2]
s = df["Duration"].value_counts().sort_values()
s = s[s > 1000]
print(s)
ax = s.plot(kind = 'bar')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')

ax.tick_params(bottom=False, left=False)

ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)
ax.set_title('Duration values', fontsize=25)
ax.set_ylabel('Count', fontsize=18)
plt.xticks(rotation=0, fontsize = 18)
plt.xlabel("")
# ax.figure.savefig("label_count.jpeg", bbox_inches="tight", dpi=1200)
```
